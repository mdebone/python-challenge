* When your tables have been merged together into one data frame, calculate the total number of wins, losses, and draws a wrestler has had over the course of their career. Also create a new column that will hold the total matches a wrestler has been in over the course of their career.

```python
# Replace all NaN values with 0 
combined_wrestler_data_clean = combined_wrestler_data2.fillna(0)
combined_wrestler_data_clean
```

```python
# Create a new column called "Total Wins" and add up each wrestler's wins per year to fill in the values
combined_wrestler_data_clean["Total Wins"] = combined_wrestler_data_clean["2013 Wins"] + combined_wrestler_data_clean["2014 Wins"] + combined_wrestler_data_clean["2015 Wins"] + combined_wrestler_data_clean["2016 Wins"]

combined_wrestler_data_clean
```

```python
# Create a new column called "Total Losses" and add up each wrestler's losses per year to fill in the values
combined_wrestler_data_clean["Total Losses"] = combined_wrestler_data_clean["2013 Losses"] + combined_wrestler_data_clean["2014 Losses"] + combined_wrestler_data_clean["2015 Losses"] + combined_wrestler_data_clean["2016 Losses"]

combined_wrestler_data_clean
```

```python
# Create a new column called "Total Draws" and add up each wrestler's draws per year to fill in the values
combined_wrestler_data_clean["Total Draws"] = combined_wrestler_data_clean["2013 Draws"] + combined_wrestler_data_clean["2014 Draws"] + combined_wrestler_data_clean["2015 Draws"] + combined_wrestler_data_clean["2016 Draws"]

combined_wrestler_data_clean
```

```python
# Create a new dataframe for those wrestlers who have wrestled at least 100 matches,
# have at least one win in 2013,
# and have at least one win in 2016
wrestler_specifc =  combined_wrestler_data_clean.loc[(combined_wrestler_data_clean["Total Matches"] >= 100) &
                                                     (combined_wrestler_data_clean["2013 Wins"] >= 1) &
                                                     (combined_wrestler_data_clean["2016 Wins"] >= 1)]

wrestler_specifc.head()
# Set the index of this new dataframe to be the wrestlers names
cream_of_the_crop = wrestler_specifc.set_index("Wrestler")

cream_of_the_crop.head()
```

```python
# Create a list of the years that we will use as our x axis
years = ["2013", "2014", "2015", "2016"]

# Plot our line that will be used to track a wrestler's wins over the years
plt.plot(years, win_history, color="green", label="Wins")

# Plot our line that will be used to track a wrestler's losses over the years
plt.plot(years, loss_history, color="red", label="Losses")


# Place a legend on the chart in what matplotlib believes to be the "best" location
plt.legend(loc="best")

plt.title(wrestler_name + "'s Recent Career")
plt.xlabel("Years")
plt.ylabel("Number of Wins/Losses")

# Print our chart to the screen
plt.show()
```
